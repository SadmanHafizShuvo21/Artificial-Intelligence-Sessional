```python

df.drop(columns=["Cabin","Name","PassengerId","Ticket"],inplace=True)
```

```python
df.head(5)
```

```python

sex = pd.get_dummies(df["Sex"],dtype=int)
Embarked = pd.get_dummies(df["Embarked"],dtype=int)
```

```python
sex
```

```python
Embarked
```

```python

df = pd.concat([df,sex],axis=1)
df = pd.concat([Embarked,df],axis=1)
df.head(5)
```

```python
df.drop(columns=["Sex","Embarked"],inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
age = df["Age"].mean()
```

```python
df["Age"].fillna(value=age)
```

```python
df["Age"] = df["Age"].fillna(value=age)
```

```python
df.isnull().sum()
```

```python
df
```

```python


from sklearn.preprocessing  import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df[["Age"]])
```

```python

df["Age"] = scaler.transform(df[["Age"]])
```

```python

fare = MinMaxScaler()
fare.fit(df[["Fare"]])
```

```python
df["Fare"] = fare.transform(df[["Fare"]])
```

```python
df
```

```python


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=["Survived"]),
                                                 df["Survived"],
                                                 test_size=20,
                                                 random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python

from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(verbose=True,max_iter=100)

```

```python
lor.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = lor.predict(X_test)
accuracy_score(y_test,y_pred)*100
```

# `With neural network:`

<br>



```python
X_train.shape
```

```python

import tensorflow as tf 
from keras import Sequential 
from keras.layers import Dense 
from keras.optimizers import Adam 
from keras.activations import relu,softmax 
from keras.losses import binary_crossentropy
```

```python


model = Sequential()
model.add(Dense(units=32,input_dim=10,activation=relu))
model.add(Dense(units=16,activation=relu))
model.add(Dense(units=8,activation=relu))
model.add(Dense(units=1,activation='sigmoid'))

model.summary()
```

```python


model.compile(optimizer=Adam(),loss=binary_crossentropy,metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100)

```

```python

y_pred = model.predict(X_test)
print(y_pred[:5])
y_pred = np.where(y_pred>0.5,1,0)
y_pred
```

```python
accuracy_score(y_test,y_pred)*100
```

```python

y_pred = model.predict(X_train)
y_pred = np.where(y_pred>0.5,1,0)
accuracy_score(y_train,y_pred)*100
```
