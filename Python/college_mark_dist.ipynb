```python

from sklearn.preprocessing import MinMaxScaler

ssc_marks = MinMaxScaler()
hsc_marks = MinMaxScaler()
college_marks = MinMaxScaler()

ssc_marks.fit(df[["SSC_Marks"]])
hsc_marks.fit(df[["HSC_Marks"]])
college_marks.fit(df[["College_Marks"]])


df["HSC_Marks"] = hsc_marks.transform(df[["HSC_Marks"]])
df["SSC_Marks"] = ssc_marks.transform(df[["SSC_Marks"]])
df["College_Marks"] = college_marks.transform(df[["College_Marks"]])

```

```python
df["Attendance_Percentage"] = df["Attendance_Percentage"]/100
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=["Grade"]),
                                                 df["Grade"],
                                                 test_size=20,
                                                 random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
from keras.utils import to_categorical

# one-hot encode target
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
y_train
```

```python
import tensorflow as tf 
from keras import Sequential 
from keras.layers import Dense 
from keras.optimizers import Adam 
from keras.activations import relu,softmax 
from keras.losses import categorical_crossentropy
```

```python


model = Sequential()
model.add(Dense(units=16,input_dim=4,activation='relu'))
model.add(Dense(units=32,activation='relu'))
model.add(Dense(units=16,activation='relu'))
model.add(Dense(units=8,activation='relu'))
model.add(Dense(units=5,activation=softmax))
model.summary()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python


model.compile(Adam(),loss=categorical_crossentropy,metrics=["accuracy"])
model.fit(X_train,y_train,epochs=200)
```
